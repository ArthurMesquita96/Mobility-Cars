# Imports

```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time
import pandas as pd
```

# Load Dataset

```python
data_full = pd.read_csv('Tabela de Pre√ßos.csv')
```

# Google Webscrapping

```python
servico = Service(ChromeDriverManager().install())
navegador = webdriver.Chrome(service=servico)
```

```python
data_cars = data_full[['marca','modelo']].drop_duplicates(ignore_index=True)
```

```python
data_cars.shape
```

```python
link_list = []
for i in range(len(data_cars)):
    marca = data_cars.loc[i, 'marca']
    modelo = data_cars.loc[i, 'modelo']
    descricao = 'fundo branco'
    print(i)

    url = 'https://www.google.com/search?q='+ marca + ' ' + modelo + ' ' +  descricao +'&tbm=isch&ved=2ahUKEwj7h6Sq3LiAAxXNAbkGHXUmAdMQ2-cCegQIABAA&oq=chevrolet+s10&gs_lcp=CgNpbWcQAzIKCAAQigUQsQMQQzIFCAAQgAQyBQgAEIAEMgUIABCABDIFCAAQgAQyBQgAEIAEMgUIABCABDIFCAAQgAQyBQgAEIAEMgcIABCKBRBDUABYAGCXBmgAcAB4AIAB6gSIAeoEkgEDNS0xmAEAqgELZ3dzLXdpei1pbWfAAQE&sclient=img&ei=xorHZPuMFc2D5OUP9cyEmA0&bih=707&biw=1536'
    navegador.get(url)
    
    time.sleep(2)

    navegador.find_element(By.CLASS_NAME, "Q4LuWd").click()
    
    time.sleep(2)
    
    try:
        data_cars.loc[i, 'link'] = navegador.find_elements(By.XPATH,"//img[contains(@class,'iPVvYb')]")[0].get_attribute('src')
    except:
        time.sleep(2)
        try:
            data_cars.loc[i, 'link'] = navegador.find_elements(By.XPATH,"//img[contains(@class,'iPVvYb')]")[0].get_attribute('src')
        except:
            data_cars.loc[i, 'link']=''
    
```
